```python
import pandas as pd
import enchant
import re
import preprocessor as p
from nltk import word_tokenize
```

```python
training = pd.read_csv('clean_training.csv', encoding='ISO-8859-1')
testing = pd.read_csv('clean_testing.csv', encoding='ISO-8859-1')
```

```python
training.head()
```

```python
# row wise function to remove URLs and emojis. takes a row and a column
def remove_urls(row, col):
    # set tweet preprocessing to remove emojis and URLs
    p.set_options(p.OPT.URL, p.OPT.EMOJI, p.OPT.HASHTAG, p.OPT.MENTION, p.OPT.SMILEY)
    
    # remove emojis and URLs from the column
    if type(row[col]) == str:
        return p.clean(row[col])
```

```python
training['text'] = training.apply(remove_urls, axis=1, col='text')
testing['text'] = testing.apply(remove_urls, axis=1, col='text')
```

```python
training = training.dropna()
```

```python
training['text'] = training['text'].str.lower()
testing['text'] = testing['text'].str.lower()
```

```python
d = enchant.Dict("en_US")
non_dict_words_train = list(set([word.encode('ascii', 'ignore') for word in word_tokenize(' '.join(training['text'])) if d.check(word) is False and re.match('^[a-zA-Z ]*$',word)] ))
non_dict_words_train
```
