```python
import pandas as pd
import numpy as np
```

```python
nb_stop = pd.read_csv('nb_stop_scores.csv')
nb_lemm = pd.read_csv('nb_lemm_scores.csv')
lr_stop = pd.read_csv('lr_stop_scores.csv')
lr_lemm = pd.read_csv('lr_lemm_scores.csv')
svc_stop = pd.read_csv('svc_stop_scores.csv')
svc_lemm = pd.read_csv('svc_lemm_scores.csv')
rf_stop = pd.read_csv('rf_stop_scores.csv')
rf_lemm = pd.read_csv('rf_lemm_scores.csv')
```

```python
nb_stop.drop('Unnamed: 0', 1, inplace=True)
nb_lemm.drop('Unnamed: 0', 1, inplace=True)
lr_stop.drop('Unnamed: 0', 1, inplace=True)
lr_lemm.drop('Unnamed: 0', 1, inplace=True)
svc_stop.drop('Unnamed: 0', 1, inplace=True)
svc_lemm.drop('Unnamed: 0', 1, inplace=True)
rf_stop.drop('Unnamed: 0', 1, inplace=True)
rf_lemm.drop('Unnamed: 0', 1, inplace=True)
```

```python
scores = pd.concat([nb_stop, nb_lemm, lr_stop, lr_lemm, svc_stop, svc_lemm, rf_stop, rf_lemm])
```

```python
scores.columns
```

```python
scores = scores[['name','acc_train','acc_test','precision_test','recall_test','roc_acc','fn_test','fp_test','acc_val','precision_val','recall_val','roc_val','fn_val','fp_val']]
```

```python
scores.head()
```

```python
scores.columns = ['name','accuracy_train','accuracy_test','precision_test','recall_test','roc_test','fn_test','fp_test','accuracy_val','precision_val','recall_val','roc_val','fn_val','fp_val']
```

```python
scores.reset_index(inplace=True)
scores.drop('index', 1, inplace=True)
```

```python
scores
```

```python
acc_train_max = scores['accuracy_train'].max()
acc_test_max = scores['accuracy_test'].max()
precision_test_max = scores['precision_test'].max()
recall_test_max = scores['recall_test'].max()
roc_test_max = scores['roc_test'].max()
fn_test_min = scores['fn_test'].min()
fp_test_min = scores['fp_test'].min()
accuracy_val_max = scores['accuracy_val'].max()
precision_val_max = scores['precision_val'].max()
recall_val_max = scores['recall_val'].max()
roc_val_max = scores['roc_val'].max()
fn_val_min = scores['fn_val'].min()
fp_val_min = scores['fp_val'].min()
```

```python
def min_max_score(row):
    count = 0
    if row['accuracy_train'] == acc_train_max:
        count += 1
    if row['accuracy_test'] == acc_test_max:
        count += 1
    if row['precision_test'] == precision_test_max:
        count += 1
    if row['recall_test'] == recall_test_max:
        count += 1
    if row['roc_test'] == roc_test_max:
        count += 1
    if row['fn_test'] == fn_test_min:
        count += 1
    if row['fp_test'] == fp_test_min:
        count += 1
    if row['accuracy_val'] == accuracy_val_max:
        count += 1
    if row['precision_val'] == precision_val_max:
        count += 1
    if row['recall_val'] == recall_val_max:
        count += 1
    if row['roc_val'] == roc_val_max:
        count += 1
    if row['fn_val'] == fn_val_min:
        count += 1
    if row['fp_val'] == fp_val_min:
        count += 1
    
    return count

scores['score'] = scores.apply(min_max_score, 1)
```

```python
scores.sort_values(by='score', ascending=False)
```

```python
scores.to_csv('final_scores.csv')
```
